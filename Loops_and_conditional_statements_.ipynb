## LOOPS IN PYTHON

**In Python, loops are used to repeat a block of code multiple times. There are two main types of loops in Python: the "for loop" and the "while loop."**

Imagine you have a basket full of colorful balls, and you want to count how many balls are in the basket. You don't want to count each ball individually because that would take a long time. So you decide to use a loop!

**For Loop**:
A for loop is like having a friend help you count the balls. You tell your friend, "Please count the balls one by one and tell me how many there are." Your friend starts counting from 1 and keeps going until they reach the last ball. Every time they count a ball, they tell you the number. This helps you keep track of the total number of balls in the basket without counting them yourself.

**While Loop**:
A while loop is a bit different. This time, you decide to count the balls yourself, but you use a special rule to help you. You start at 1 and say, "While there are still balls in the basket, keep counting and tell me the number." So you pick up a ball, say "1," and put it aside. Then you check if there are more balls left in the basket. If there are, you pick up the next ball, say "2," and put it aside. You keep doing this until there are no more balls in the basket.

In both cases, the loop helps you repeat a task (counting the balls) until a specific condition is met (reaching the last ball). The for loop knows how many times it needs to count because you tell it the total number of balls. The while loop keeps counting as long as there are still balls in the basket.

So, loops are like helpful instructions that let you repeat an action without doing it over and over again by yourself. They make tasks easier and more efficient, just like having a friend or a special rule to help you count the balls!

```
#Check if a number is prime
num = int(input("Enteranumber:"))
is_prime = True
for i in range(2, num):
  if num % i == 0:
    is_prime = False  #numberisnotprime
    break    #exitfromforloop

if is_prime:
  print(num, "isprime")
else:
  print(num, "isnotaprime")
```
