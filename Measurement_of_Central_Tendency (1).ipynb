# 1. Mean:
**Definition**:
The mean is the average of a set of values. It is calculated by adding up all the values and dividing by the number of values.

**Real-Life Example**:
Consider the exam scores of a class. The mean score gives an idea of the average performance.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
scores = [85, 90, 92, 88, 95, 78, 89, 92, 93, 87]

# Calculate mean
mean_score = np.mean(scores)

# Plotting
plt.hist(scores, bins=10, edgecolor='black')
plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_score:.2f}')
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Mean Calculation in Python')
plt.legend()
plt.show()
```

# 2. Mode:
**Definition**:
The mode is the value that appears most frequently in a dataset.

**Real-Life Example**:
In a survey, the mode of the responses indicates the most common answer.

```python
from scipy.stats import mode
import matplotlib.pyplot as plt

# Sample data
ages = [22, 25, 28, 22, 30, 25, 22, 28, 32, 28]

# Calculate mode
mode_result = mode(ages)
modes = mode_result.mode  # Access the mode attribute

# Plotting
plt.hist(ages, bins=8, edgecolor='black')

# Check if there are multiple modes
if isinstance(modes, np.ndarray):
    for mode_value in modes:
        plt.axvline(mode_value, color='green', linestyle='dashed', linewidth=2, label=f'Mode: {mode_value}')
else:
    plt.axvline(modes, color='green', linestyle='dashed', linewidth=2, label=f'Mode: {modes}')

plt.xlabel('Ages')
plt.ylabel('Frequency')
plt.title('Mode Calculation in Python')
plt.legend()
plt.show()
```

# 3. Median:
**Definition**:
The median is the middle value in a dataset when it is sorted in ascending order.

**Real-Life Example** :
In a salary dataset, the median salary represents the middle value, unaffected by extreme values.

```python
# Sample data
salaries = [50000, 55000, 60000, 62000, 63000, 70000, 75000, 80000, 85000, 90000]

# Calculate median
median_salary = np.median(salaries)

# Plotting
plt.hist(salaries, bins=6, edgecolor='black')
plt.axvline(median_salary, color='blue', linestyle='dashed', linewidth=2, label=f'Median: {median_salary}')
plt.xlabel('Salaries')
plt.ylabel('Frequency')
plt.title('Median Calculation in Python')
plt.legend()
plt.show()
```

# Range:
**Definition**:The range is the difference between the maximum and minimum values in a dataset.

**Real-Life Example**:
For a set of daily temperatures, the range is the difference between the highest and lowest temperatures.

```python
temperature_readings = [68, 72, 75, 60, 80]
temperature_range = max(temperature_readings)-min(temperature_readings)

print(f"Temperature Range: {temperature_range}")
```

**Here are the top 5 interview questions for a Machine Learning Engineer related to the concepts of central tendency (mean, mode, median)=**


# 1. Explain the Importance of Central Tendency in Data Analysis.
**Purpose**: This question assesses the candidate's understanding of why central tendency measures (mean, mode, median) are crucial in data analysis.
**Expected Answer**: The candidate should explain that central tendency measures provide a summary of the central or average value in a dataset, helping to understand the typical or representative value. These measures are fundamental for making statistical inferences and gaining insights into the distribution of data.
# 2. When Would You Prefer to Use Median Instead of Mean in Describing a Dataset?
**Purpose**: This question evaluates the candidate's ability to choose the appropriate measure of central tendency based on the characteristics of the data.
**Expected Answer**: The candidate should mention situations where the data may have outliers or is not normally distributed. In such cases, the median is a more robust measure as it is not affected by extreme values.
#3. How Would Outliers Affect the Mean, Mode, and Median?
**Purpose**: This question examines the candidate's understanding of the impact of outliers on different measures of central tendency.
**Expected Answer**: The candidate should explain that outliers can significantly influence the mean, less so the median, and minimally affect the mode. The mean is sensitive to extreme values, while the median is resistant to them.
# 4. Can You Provide an Example Where the Mean and Median Differ Significantly? What Does This Indicate?
**Purpose**: This question tests the candidate's ability to identify scenarios where the mean and median provide different insights.
**Expected Answer**: The candidate may offer examples like skewed distributions, where the mean is pulled in the direction of the skewness. This indicates that the data is not symmetrically distributed.
# 5. How Does Skewness Affect the Interpretation of Mean and Median?
**Purpose**: This question assesses the candidate's understanding of the relationship between skewness and central tendency measures.
**Expected Answer**: The candidate should explain that in a positively skewed distribution, the mean is greater than the median, while in a negatively skewed distribution, the mean is less than the median. Skewness indicates the direction and degree of asymmetry in the data.
